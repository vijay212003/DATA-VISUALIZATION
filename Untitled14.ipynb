```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load the dataset (replace with your actual file path)
# Make sure the file 'supermarket_sales.csv' is uploaded to the correct location,
# or update the path below if it is in a different location.
# To upload the file, click the folder icon on the left sidebar and then click the upload icon.
df = pd.read_csv("/content/SuperMarket Analysis.csv")

# Select important categorical columns for network analysis
# Example: Branch, Product line, Payment
branch_col = "Branch"
product_col = "Product line"
payment_col = "Payment"

# Create Graph
G = nx.Graph()

# Add nodes for Branches
for branch in df[branch_col].unique():
    G.add_node(branch, type="Branch")

# Add nodes for Product lines
for product in df[product_col].unique():
    G.add_node(product, type="Product")

# Add nodes for Payment methods
for pay in df[payment_col].unique():
    G.add_node(pay, type="Payment")

# Add edges: Branch  Product
for _, row in df.iterrows():
    G.add_edge(row[branch_col], row[product_col])
    G.add_edge(row[branch_col], row[payment_col])

# Draw the graph
plt.figure(figsize=(14, 10))
pos = nx.spring_layout(G, k=0.5, seed=42)

# Node colors by type
color_map = []
for node in G.nodes(data=True):
    if node[1]["type"] == "Branch":
        color_map.append("lightblue")
    elif node[1]["type"] == "Product":
        color_map.append("lightgreen")
    else:  # Payment
        color_map.append("orange")

nx.draw(
    G, pos, with_labels=True, node_size=1500, node_color=color_map,
    font_size=9, edge_color="gray", font_weight="bold"
)

plt.title("Retail Supermarket Sales Network (Branches - Products - Payments)", fontsize=14)
plt.show()
```
