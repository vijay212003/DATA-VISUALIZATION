```python
import pandas as pd
import plotly.express as px

# Load the data
df = pd.read_csv('/content/SuperMarket Analysis.csv')

# Generate Treemap
fig_treemap = px.treemap(
    df,
    path=["City", "Branch", "Product line"],
    values="Sales",
    color="City",
    hover_data=["Gender", "Payment"],
    title="Supermarket Sales - TreeMap"
)
fig_treemap.show()
fig_treemap.write_json("treemap.json")

# Generate Sunburst Chart
fig_sunburst = px.sunburst(
    df,
    path=["City", "Branch", "Product line"],
    values="Sales",
    color="City",
    hover_data=["Gender", "Payment"],
    title="Supermarket Sales - Sunburst Chart"
)
fig_sunburst.show()
fig_sunburst.write_json("sunburst.json")
```
