```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

# ======================
# Step 1: Load dataset
# ======================
df = pd.read_csv("SuperMarket Analysis.csv")

# Fix numeric columns (convert Sales and others if stored as strings)
for col in ["Unit price", "Quantity", "Tax 5%", "Sales", "cogs", "gross income", "Rating"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# ======================
# Step 2: Identify Columns
# ======================
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
continuous_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()

print("Categorical Columns:", categorical_cols)
print("Continuous Columns:", continuous_cols)

# ================================
# Step 3: Categorical vs Categorical
# Example: Customer type vs Gender
# ================================
cross_tab = pd.crosstab(df['Gender'], df['Customer type'])

# Stacked Bar Chart
cross_tab.plot(kind="bar", stacked=True, figsize=(6,4))
plt.title("Stacked Bar Chart: Customer type vs Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Grouped Bar Chart
cross_tab.plot(kind="bar", stacked=False, figsize=(6,4))
plt.title("Grouped Bar Chart: Customer type vs Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Segmented Bar Chart (Normalized)
(cross_tab.T / cross_tab.T.sum()).T.plot(kind="bar", stacked=True, figsize=(6,4))
plt.title("Segmented Bar Chart: Customer type vs Gender")
plt.xlabel("Gender")
plt.ylabel("Proportion")
plt.show()

# Mosaic Plot
plt.figure(figsize=(8,6))
mosaic(df, ['Gender','Customer type'])
plt.title("Mosaic Plot: Customer type vs Gender")
plt.show()

# ================================
# Step 4: Continuous vs Continuous
# Example: Unit price vs Quantity
# ================================
sns.lmplot(x="Unit price", y="Quantity", data=df, height=5, aspect=1.2, scatter_kws={'alpha':0.6})
plt.title("Scatterplot with Fit Line: Unit Price vs Quantity")
plt.show()

# ================================
# Step 5: Categorical vs Continuous
# Example: Gender vs Sales
# ================================

# Bar Chart
plt.figure(figsize=(6,4))
sns.barplot(x="Gender", y="Sales", data=df, ci="sd")
plt.title("Bar Chart: Gender vs Sales")
plt.show()

# Grouped Kernel Density Plot
plt.figure(figsize=(6,4))
sns.kdeplot(data=df, x="Sales", hue="Gender", fill=True, common_norm=False, alpha=0.5)
plt.title("Grouped Kernel Density Plot: Gender vs Sales")
plt.show()

# Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x="Gender", y="Sales", data=df)
plt.title("Box Plot: Gender vs Sales")
plt.show()

# Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(x="Gender", y="Sales", data=df, inner="quartile")
plt.title("Violin Plot: Gender vs Sales")
plt.show()

# Beeswarm Plot
plt.figure(figsize=(6,4))
sns.swarmplot(x="Gender", y="Sales", data=df, size=3)
plt.title("Beeswarm Plot: Gender vs Sales")
plt.show()
```
